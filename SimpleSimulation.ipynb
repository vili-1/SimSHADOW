```python
import io
import pandas as pd

import cirq
from cirq.contrib.qasm_import import circuit_from_qasm
from qiskit.visualization import plot_histogram

# -------------------------------
# Input QASM Code
# -------------------------------

qasm_program_template = """OPENQASM 3.0;
include "stdgates.inc";
qubit q0;
rx({t0_value}) q0;
rz({t1_value}) q0;
rz({t1_value}) q0;
"""

# -------------------------------
# HeatMap Fix
# -------------------------------

# Add what to amend <--- here we will read the heat map
parameter_value_1 = 0.52
parameter_value_2 = 0.51

# -------------------------------
# Load QASM to Platform Circuit
# -------------------------------
resolved_qasm_program = qasm_program_template.format(t0_value=parameter_value_1,t1_value=parameter_value_2)
print("--- Resolved QASM Program (Input to Cirq) ---")
print(resolved_qasm_program)
print("---------------------------------------------")
try:
    __cirq_circuit = circuit_from_qasm(resolved_qasm_program)
    print(__cirq_circuit)
except Exception as e:
    print(f"An error occurred during import: {e}")
for q in __cirq_circuit.all_qubits():
    __cirq_circuit.append(cirq.measure(q, key=f"m_{q}"))

# -------------------------------
# Simulation Flow
# -------------------------------
# Run simulation with Noise model
__sim = cirq.Simulator()
noisy_circuit = __cirq_circuit.with_noise(cirq.depolarize(p=0.0001))
__noise_result = __sim.run(noisy_circuit, repetitions=10000)

# -------------------------------
# Print Results
# -------------------------------
__counts = __noise_result.data.value_counts().to_dict()
print(">>> Measurement outcomes:", __counts)
plot_histogram(__counts)
```

```python
from qiskit import QuantumCircuit
from qiskit_qasm3_import import parse
from qiskit_aer import AerSimulator
from qiskit_aer.noise import NoiseModel, depolarizing_error
from qiskit import transpile
from qiskit.visualization import plot_histogram


# -------------------------------
# utilities
# -------------------------------
def max_gate_qubits(circ: QuantumCircuit) -> int:
    """
    Return the maximum number of qubits used by any *quantum* operation
    in the circuit (ignores classical-only ops, if any).
    """
    max_q = 0
    for inst in circ.data:
        op = inst.operation
        n = op.num_qubits
        if n > max_q:
            max_q = n
    return max_q

def get_gate_names(circ):
    names = set()
    for inst in circ.data:
        op = inst.operation
        if op.name not in {"measure", "reset"}:   # skip non-unitary ops
            names.add(op.name)
    return sorted(names)

# -------------------------------
# Input QASM Code
# -------------------------------

qasm_program_template = """OPENQASM 3.0;
include "stdgates.inc";
qubit q0;
rx({t0_value}) q0;
rz({t1_value}) q0;
rz({t0_value}) q0;
"""

# -------------------------------
# HeatMap Fix
# -------------------------------

# Add what to amend <--- here we will read the heat map
parameter_value_1 = 0.52
parameter_value_2 = 0.51

# -------------------------------
# Load QASM to Platform Circuit
# -------------------------------
resolved_qasm_program = qasm_program_template.format(t0_value=parameter_value_1,t1_value=parameter_value_2)
print("--- Resolved QASM Program (Input to Qiskit) ---")
print(resolved_qasm_program)
print("----------------------------------------------")
try:
    # Parse the QASM3 text code
    qasm3_ast = parse(resolved_qasm_program)
    if qasm3_ast is None:
        print("Error: AST is empty!")

    circuit: QuantumCircuit = qasm3_ast
    print(circuit.draw())
except Exception as e:
    print(f"An error occurred during import: {e}")
circuit.measure_all()

# -------------------------------
# Simulation Flow
# -------------------------------

# --- 1. Adding Noise model
noise_model = NoiseModel()
max = max_gate_qubits(circuit)
for i in range(1, max+1):
    error = depolarizing_error(0.0001, i)
gate_list = list(get_gate_names(circuit))
noise_model.add_all_qubit_quantum_error(error, gate_list)
# --- 2. Create a noisy simulator ---
sim = AerSimulator(noise_model=noise_model)
# --- 3. Compile your circuit for this simulator ---
compiled_circuit_noise = transpile(circuit, sim)
# --- 4. Run simulation with Noise model
job = sim.run(compiled_circuit_noise, shots=10000)
print(compiled_circuit_noise)

# -------------------------------
# Print Results
# -------------------------------
counts = job.result().get_counts()
print("Measurement outcomes:", counts)
plot_histogram(counts)
```

```python
import json
import zipfile
import numpy as np
from pathlib import Path
import json
import os
import glob
import pprint
import matplotlib.pyplot as plt
import numpy as np

def plot_results(arr,title,_vmax=2.0,_vmin=-2.0,colour="coolwarm"):

  # 15 observables
  observables = ["XX","XY","XZ","YX","YY","YZ","ZX","ZY","ZZ","XI","IX","YI","IY","ZI","IZ"]
  # 9 states
  states = ["|00⟩","|01⟩","|10⟩","|11⟩","|+0⟩","|-0⟩","|0+⟩","|0-⟩","|Φ⁺⟩"]

  fig, ax = plt.subplots(figsize=(9, 4.5))
  #im = ax.imshow(arr, cmap="coolwarm", aspect="auto", interpolation="nearest")
  im = ax.imshow(
    arr,
    cmap=colour,
    aspect="auto",
    interpolation="nearest",
    vmin=_vmax,
    vmax=_vmin
  )


  # Axes
  ax.set_xticks(np.arange(len(observables)))
  ax.set_yticks(np.arange(len(states)))
  ax.set_xticklabels(observables, rotation=45, ha="right", fontsize=9)
  ax.set_yticklabels(states, fontsize=10)

  # Labels
  ax.set_xlabel("Observable", fontsize=12)
  ax.set_ylabel("State", fontsize=12)
  ax.set_title(title, fontsize=14, weight='bold')

  # Colorbar
  cbar = fig.colorbar(im, ax=ax)
  cbar.set_label("E_noisy - E_ideal", rotation=270, labelpad=15)

  # Uncommnet if you want to see the actual valudes:
  #fmt="{:.5f}"
  #for i in range(arr.shape[0]):
  #  for j in range(arr.shape[1]):
  #      val = arr[i, j]
  #      ax.text(
  #          j, i,
  #          fmt.format(val),
  #          ha="center",
  #          va="center",
  #          fontsize=7,
  #          color="black"
  #      )

  for i in range(arr.shape[0]):
    for j in range(arr.shape[1]):
        val = arr[i, j]

        # choose what you want to show for (near-)zero values
        if val > 0:
            s = "+"
        elif val < 0:
            s = "-"
        else:
            s = ""      # or "0" or "·"

        ax.text(
            j, i,
            s,
            ha="center",
            va="center",
            fontsize=7,
            color="yellow"
        )

  plt.tight_layout()
  plt.show()

path = '/content/results/' #SimSHADOW_artifact/results/'
fingerprints = {}
fingerprintstypes = []

for json_filename in glob.glob(os.path.join(path, '*.json')):
  # Fixed JSON file name (from inspection)
  #json_filename = "simshadow_results_20251215_181631.json"
  #id="20251215_181631"
  id = Path(json_filename).stem.replace("simshadow_results_", "", 1)
  with open(json_filename, 'r') as file:
      data = json.load(file)

  # Extract fingerprint_data section
  fingerprint_data = data["fingerprint_data"]
  # Convert each [platform][noise_type] matrix into a numpy array
  for platform, noises in fingerprint_data.items():
      for noise_type, matrix in noises.items():
          key = f"{platform}/{noise_type}/{id}"
          fingerprints[key] = np.array(matrix)
          item = f"{platform}/{noise_type}"
          if item not in fingerprintstypes:
            fingerprintstypes.append(item)

debug = False
if (debug):
  # Just to test
  for key in fingerprintstypes:
    print(key)
  for inner_key, arr in fingerprints.items():
    print(f"{inner_key}: shape={arr.shape}")
    print(arr, "\n")

# --- Group matrices by type ---
grouped = {ftype: [] for ftype in fingerprintstypes}
for key_full, arr in fingerprints.items():
    key_type = "/".join(key_full.split("/")[:2])  # strip the run_id
    grouped[key_type].append(arr)
# --- Compute elementwise mean and std for each type ---
mean_std = {}
for ftype, matrices in grouped.items():
    stack = np.stack(matrices, axis=0)   # shape (N, 9, 15)
    mean_std[ftype] = {
        "mean": np.mean(stack, axis=0),
        "std": np.std(stack, axis=0),
        "n": stack.shape[0]
    }
# --- Get the global min/max to set it for the graphs uniqely
mean_min = min(np.min(stats["mean"]) for stats in mean_std.values())
mean_max = max(np.max(stats["mean"]) for stats in mean_std.values())
if mean_min < 0:
    abs_mean_max = max(mean_max, -mean_min)
else:
    abs_mean_max = max(mean_max, 0.0)

std_min  = min(np.min(stats["std"])  for stats in mean_std.values())
std_max  = max(np.max(stats["std"])  for stats in mean_std.values())
if std_min < 0:
    abs_std_max = max(std_max, -std_min)
else:
    abs_std_max = max(std_max, 0.0)

import numpy as np
import matplotlib as mpl
from matplotlib.colors import ListedColormap

base = mpl.cm.get_cmap("twilight_shifted")
# choose a window that avoids the wraparound endpoints
colors = base(np.linspace(0.08, 0.92, 256))
cmap = ListedColormap(colors, name="twilight_shifted_trunc")

# --- Display results ---
for ftype, stats in mean_std.items():
    if (debug):
      print(f"\n=== {ftype} === (N={stats['n']})")
      print("Elementwise mean (9x15):")
      print(stats["mean"])
      print("\nElementwise std (9x15):")
      print(stats["std"])
    plot_results(stats["mean"],ftype+": MEAN",abs_mean_max*1.02,-abs_mean_max*1.02,cmap)
    plot_results(stats["std"],ftype+": STDV",abs_std_max*1.02,-abs_std_max*1.02,"PiYG")
```
